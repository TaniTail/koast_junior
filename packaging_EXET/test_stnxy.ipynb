3. 기존 파일과 비교 후 업데이트

   2. 파일이 이미 존재하며 지점정보가 완전히 같은 경우 : 비교 후 스킵

```python
#2. 파일이 이미 존재하며 지점정보가 완전히 같은 경우 : 비교 후 스킵

# 19930304 AWS 관측지점 로드
obscode = 'aws' # aws,asos,gts
modelcode = 'gdps_ne36' #gdps_ne36, klfs_ne36
strDt = '19930304'
dt = datetime.strptime(strDt, '%Y%m%d')

points = stn.reqApiAwsInfoData(dt)
print(type(points))
print(points)

# 19930304 관측지점에 따라 격자점 인덱스 정보 추출후 저장


modelStnInfos = stn.model_station_info(dt, obscode, modelcode)
stn.writeXyCsv(obscode, modelcode, modelStnInfos, dt)

```

```python
# 비교
xyfile = cfg.PATH_MODEL_STNXY.format(MODEL=modelcode, OBS=obscode)

if not os.path.exists(xyfile): 
                
    print(f"File does not exist: {xyfile}, file writing.")
    modelStnInfos= stn.model_station_info(dt, obscode, modelcode)
    stn.writeXyCsv(obscode, modelcode, modelStnInfos, dt)
    
else :
    print(f"File exists: {xyfile}, comparing content.")
    
    stnApiList = stn.reqApiAwsInfoData(dt)
    # print(stnApiList)
    existingData = pd.read_csv(xyfile, header=0)
    
    # print("check",existingData)
    
    stnApiList_df = pd.DataFrame(stnApiList)
    stnApiList_df = stnApiList_df.rename(columns={'id': 'stnId', 'lat': 'lat', 'lon': 'lon'})
    stnApiList_df = stnApiList_df[['stnId', 'lat', 'lon']] 
    
    print('comparing dataframes')
    
    
    print(existingData.head())
    print(stnApiList_df.head())
    
    existingData['stnId'] = existingData['stnId'].astype(int) # 여기서 속도가 좀 많이 느려지는 것
    stnApiList_df['stnId'] = stnApiList_df['stnId'].astype(int)
    
    merged_df = pd.merge(existingData[['stnId', 'lat', 'lon']], 
                                    stnApiList_df, 
                                    on='stnId', 
                                    how='outer',  # 모든 stnId를 포함
                                    suffixes=('_existing', '_api'))
    
    changes_detected = merged_df[(merged_df['lat_existing'] != merged_df['lat_api']) | 
                                            (merged_df['lon_existing'] != merged_df['lon_api'])]
    
    print("print differences : ",changes_detected)
    
    if not changes_detected.empty:
        print(f"Changes detected, updating file: {xyfile}")
        modelStnInfos= stn.model_station_info(dt, obscode, modelcode)
        stn.writeXyCsv(obscode, modelcode, modelStnInfos, dt)
    else:
        print(f"No changes detected, skipping file update: {xyfile}")
    
```

4. 더미파일 생성 (내용 랜덤)

```python
import random

# 파일 생성 함수
def create_dummy_file(file_name, date_str, num_lines=7, num_values_per_line=25): # 왜 25개인지 확인할 것 
    with open(file_name, 'w') as file:
        # 헤더 작성 (파일의 날짜 반영)
        ymdh = f"ymdh:{date_str}00\n"
        file.write(f"# INFO, fcstInterval:3, fcstMaxHour:120, {ymdh}")
        
        # 랜덤 데이터 생성
        for _ in range(num_lines):
            # 랜덤한 정수 ID (6자리)
            stn_id = random.randint(100000, 999999) # id 생성이 랜덤이어도 되나?
            
            # 25개의 랜덤 부동 소수점 숫자 생성 (0.0에서 3.5 사이)
            values = [f"{random.uniform(0, 3.5):.1f}" for _ in range(num_values_per_line)]
            
            # ID와 값들을 한 줄로 작성
            line = f"{stn_id}  " + "      ".join(values) + "\n"
            file.write(line)


obscode = 'aws' # aws,asos,gts
modelcode = 'gdps_ne36' #gdps_ne36, klfs_ne36



# 날짜 리스트 생성 (예: 10일간의 파일 생성)
start_date = datetime(2024, 1, 11)  # 기준 날짜
for i in range(10):  # 10개의 파일을 생성
    # 날짜 계산
    current_date = start_date + timedelta(days=i)
    date_str = current_date.strftime('%Y%m%d')
    
    # 파일 이름과 날짜에 맞춘 더미 파일 생성
    file_name = f"extract_{modelcode}_{obscode}.{date_str}12" # extract_gdps_ne36_asos.2023101400
    create_dummy_file(file_name, date_str)
    print(f"{file_name} 생성 완료.")
```
